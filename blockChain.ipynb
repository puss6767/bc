# Hash

```python
import hashlib

# Step 1: Input string
message = "Blockchain Developer"

# Step 2: Convert to bytes (important!)
message_bytes = message.encode()

# Step 3: Apply SHA-256
hash_object = hashlib.sha256(message_bytes)

# Step 4: Get hexadecimal digest
digest = hash_object.hexdigest()

# Step 5: Print result
print("Original Message:", message)
print("SHA-256 Digest:", digest)
```

```python
import hashlib
import time

class Block:
    def __init__(self, data, previous_hash):
        self.data = data
        self.previous_hash = previous_hash
        self.timestamp = time.time()
        self.nonce = 0
        self.hash = self.mine_block()

    def calculate_hash(self):
        block_string = str(self.data) + str(self.previous_hash) + str(self.timestamp) + str(self.nonce)
        return hashlib.sha256(block_string.encode()).hexdigest()

    def mine_block(self, difficulty=4):
        prefix = "0" * difficulty

        while True:
            hash_value = self.calculate_hash()
            if hash_value.startswith(prefix):
                return hash_value
            else:
                self.nonce += 1


# -----------------------------
# Create Blockchain

blockchain = []

# Genesis block
genesis = Block("Genesis Block", "0")
blockchain.append(genesis)

# Add more blocks
block1 = Block("Transaction 1", blockchain[-1].hash)
blockchain.append(block1)

block2 = Block("Transaction 2", blockchain[-1].hash)
blockchain.append(block2)

# -----------------------------
# Print blockchain

for i, block in enumerate(blockchain):
    print(f"\nBlock {i}")
    print("Data:", block.data)
    print("Hash:", block.hash)
    print("Previous Hash:", block.previous_hash)
```

# Signing Document


```python
import hashlib

# -----------------------------
# RSA Keys (same as before)

p = 61
q = 53
n = p * q
phi = (p - 1) * (q - 1)
e = 17

# Find d
d = 1
while (d * e) % phi != 1:
    d += 1

print("Public Key:", (e, n))
print("Private Key:", (d, n))

# -----------------------------
# Message

message = "Hello World"

# Step 1: Hash message
hash_value = hashlib.sha256(message.encode()).hexdigest()

# Convert hash to integer (important!)
hash_int = int(hash_value, 16)

# -----------------------------
# Step 2: Sign (encrypt hash with private key)

signature = (hash_int ** d) % n

print("\nSignature:", signature)

# -----------------------------
# Step 3: Verify

# Decrypt signature using public key
decrypted_hash = (signature ** e) % n

# Hash message again
new_hash = int(hashlib.sha256(message.encode()).hexdigest(), 16)

# Compare
if decrypted_hash == new_hash:
    print("✅ Signature Verified")
else:
    print("❌ Signature Invalid")
```

# Block Hash


```python
import hashlib
import time

class Block:
    def __init__(self, transactions, previous_hash):
        self.transactions = transactions  # list of transactions
        self.previous_hash = previous_hash
        self.timestamp = time.ctime()
        self.nonce = 0
        self.hash = self.mine_block()

    def calculate_hash(self):
        block_data = str(self.transactions) + str(self.previous_hash) + str(self.timestamp) + str(self.nonce)
        return hashlib.sha256(block_data.encode()).hexdigest()

    def mine_block(self, difficulty=4):
        prefix = "0" * difficulty
        while True:
            hash_value = self.calculate_hash()
            if hash_value.startswith(prefix):
                return hash_value
            self.nonce += 1


# -----------------------------
# Create Blockchain

blockchain = []

# Genesis Block
genesis_block = Block(["Genesis Transaction"], "0")
blockchain.append(genesis_block)

# Block 1
block1 = Block(
    ["A sends 10 BTC to B", "B sends 5 BTC to C"],
    blockchain[-1].hash
)
blockchain.append(block1)

# Block 2
block2 = Block(
    ["C sends 2 BTC to D"],
    blockchain[-1].hash
)
blockchain.append(block2)

# -----------------------------
# Print Blockchain

for i, block in enumerate(blockchain):
    print(f"\n🔹 Block {i}")
    print("Transactions:", block.transactions)
    print("Timestamp:", block.timestamp)
    print("Hash:", block.hash)
    print("Previous Hash:", block.previous_hash)
```

# validity

```python
import hashlib
import time

class Block:
    def __init__(self, data, previous_hash):
        self.data = data
        self.previous_hash = previous_hash
        self.timestamp = time.time()
        self.nonce = 0
        self.hash = self.mine_block()

    def calculate_hash(self):
        value = str(self.data) + str(self.previous_hash) + str(self.timestamp) + str(self.nonce)
        return hashlib.sha256(value.encode()).hexdigest()

    def mine_block(self, difficulty=3):
        prefix = "0" * difficulty
        while True:
            hash_value = self.calculate_hash()
            if hash_value.startswith(prefix):
                return hash_value
            self.nonce += 1


# -----------------------------
# Create Blockchain

blockchain = []
blockchain.append(Block("Genesis Block", "0"))

for i in range(1, 5):
    data = f"Block {i} Data"
    prev_hash = blockchain[-1].hash
    blockchain.append(Block(data, prev_hash))


# -----------------------------
# Function to check validity

def is_chain_valid(chain):
    for i in range(1, len(chain)):
        current = chain[i]
        previous = chain[i - 1]

        # 1. Check current hash
        if current.hash != current.calculate_hash():
            print(f"❌ Block {i} hash is invalid")
            return False

        # 2. Check previous hash link
        if current.previous_hash != previous.hash:
            print(f"❌ Block {i} previous hash mismatch")
            return False

    return True


# -----------------------------
# Check validity

print("Is blockchain valid?", is_chain_valid(blockchain))


# -----------------------------
# Tamper with blockchain

print("\n⚠️ Tampering with Block 2...\n")
blockchain[2].data = "Hacked Data"

# Check again
print("Is blockchain valid?", is_chain_valid(blockchain))
```
